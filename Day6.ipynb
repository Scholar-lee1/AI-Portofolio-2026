```python
import datetime
import math

# Create a class for the game
class BirthdayCountdown:
    def __init__(self, name, birth_month, birth_day):
        self.name = name
        self.birth_month = birth_month
        self.birth_day = birth_day

    def days_until_birthday(self):
        today = datetime.date.today()
        current_year = today.year
        birthday = datetime.date(current_year, self.birth_month, self.birth_day)

        #If birthday already passed this year, check next year
        if birthday < today:
            birthday = datetime.date(current_year + 1, self.birth_month, self.birth_day)

        days_left = (birthday - today).days
        return days_left


# Ask the player for their name and birthday
name = input("What is your name?")
birth_month = int(input("Enter your birth month (1-12)"))
birth_day = int(input("Enter your birth month (1-31)"))

# Create the Player
player = BirthdayCountdown(name, birth_month, birth_day)

# Get days left
days = player.days_until_birthday()

# Fun math: Calculate weeks too
weeks = math.ceil(days/7)

# Show the result
print(f"Hi {player.name}! Your birthday is in {days} day!")
print(f"That's about {weeks} weeks!")
```
