```python
# Student score analyser

# Problem: Analyse exam performance of students

# Features:

# Generate random scores for students

# Find average, top, and bottom performeers

# Normalize the scores(0-100 scale)
```

```python
import numpy as np

scores = np.random.randint(40, 100, size=(10,))
norm_scores = (scores - scores.min())/ (scores.max() - scores.min()) * 100
print("Normalized Scores:", norm_scores)
```

```python
#  Temperature Data Tracker

# Problems: Analyze weekly temperature data from sensors

# Features:

# Record Temperature for a week

# Find hottest, coldest days

# Detect temperature spike (diff > 0)
```

```python
temps = np.array([30,31, 35, 32, 33, 38, 36])
print("Spike Detected:", np.where(np.diff(temps) > 2))
```
