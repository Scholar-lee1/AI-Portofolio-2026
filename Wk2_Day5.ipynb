```python
# Project Brief: Clean the data for machine learning
# Common issues in dataset

#    Mixed-case and inconsistent value(e.g "male", "MALE", "FeMale")
#    Invalid email address
#    Inconsistent subscription plan ("Premium" vs "premium")
#    String values that need encoding for ML (e.g. "Charmed": Yes/No)
```

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/AI-2026/Dirty_Customer_Data.xlsx'

# Read the data
df = pd.read_excel(file_path)

# Step 2: Remove duplicates
df = df.drop_duplicates()

# Step 3: Standardize Gender
df["Gender"] = df["Gender"].str.strip().str.capitalize()
df["Gender"] = df["Gender"].replace({"FeMale": "Female"})

# Step 4: Fix Email column
df = df[df["Email"].str.contains("@")]

# Step 5: Normalize Subscription plan
df["SubscriptionPlan"] = df["SubscriptionPlan"].str.strip().str.capitalize()

# Step 6: Fill in the missing values
df["Age"] = df["Age"].fillna(df["Age"].median())
df["DailyUsageHours"] = df["DailyUsageHours"].fillna(df["DailyUsageHours"].mean())
df["SubscriptionPlan"] = df["SubscriptionPlan"].fillna("Free")
df["Churned"] = df["Churned"].fillna("No")

# Step 7:Normalize churned values
df["Churned"] = df["Churned"].str.strip().str.lower().map({"yes": 1, "no" : 0})

# Step 8: Convert SignupDate to Timeline
df["SignupDate"] = pd.to_datetime(df["SignupDate"])

# Now (it's clean and ready for ML(e.g training a churn prediction model))
print((df.head()))
```
