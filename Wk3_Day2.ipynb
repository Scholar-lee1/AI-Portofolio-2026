```python
# Matplotlib is a comprehensive library used for creating static,
# animated, and interactive visualization in python. It is commonly
# used for data exploration and presentation, and work seamlessly with
# NumPy and Pandas
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
# Bar Chart

plt.bar(['A', 'B', 'C'], [5, 7, 3])
plt.title("Bar Chart")
plt.show()
```

```python
# Scatter Plot

plt.scatter([1, 2, 3], [3, 1, 5])
plt.title("Scatter Plot")
plt.show()
```

```python
# Pie Chart

plt.pie([30, 20, 50], labels = ["Apples", "Bananas", "Cherries"], autopct = '%1.1f%%' )
plt.title("Fruit Distribution")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

plt.plot(x,y, color='red', linestyle='--', marker='o', linewidth=2)
plt.title("Customized line plot")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.plot(x, y1, label="Ascending")
plt.plot(x,y2,label="Descending")
plt.legend()
plt.title("Two lines")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 4))

axs[0].plot((1,2,3), [3, 2, 1])
axs[0].set_title("Left Plot")

axs[1].bar([1, 2, 3], [3, 2, 5])
axs[1].set_title("Right Plot")

plt.tight_layout()
plt.show()
```
