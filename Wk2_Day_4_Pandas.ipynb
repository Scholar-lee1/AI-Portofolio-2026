```python
# Attendance & Productivity Tracker

# Task: Manage and analyze class or work attendance


# What they will Do:

#    Create a DataFrame with names, dates, login/logout times (or attendance status)

#    Calculate total hours present (if using login/logout).

#    Count days present versus absent

#    lag students/employees with less than 75% attendance.

# Real life Relevance: Mimics school/work attendance systems. Helps them see how performance can be tracked
```

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/AI-2026/Attendance_Tracker.xlsx'

# Read the data
df = pd.read_excel(file_path)
df.head()

# Show the first few rows
print(df.head())

# Count total number of records for each student
total_days = df.groupby("Name")["Status"].count()

# Count how many days each student was present
present_days = df[df["Status"] == "Present"].groupby("Name")["Status"].count()

# Combine and calculate attendance pecentage
attendance_summary = pd.DataFrame({
    "Total Days": total_days,
    "Days Present": present_days,
})

attendance_summary["Attendance %"] = (attendance_summary["Days Present"] / attendance_summary["Total Days"]) * 100

print("\nAttendance Summary:")
print(attendance_summary.sort_values(by ="Attendance %", ascending = False))

# Flag those below 75%
print("\nStudent with low Attendance")
print(attendance_summary[attendance_summary["Attendance %"] < 75])
```
