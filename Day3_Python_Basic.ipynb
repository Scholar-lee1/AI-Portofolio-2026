```python
#  Set the secret number
secret_number = 7

# Give the player 3 chance to guess (for loop)
for attempt in range(3):
    guess = int(input("Guess the secret number between one and 10: "))
    if guess == secret_number:
        print("You guessed it right! You win!")
        break
    else:
        print("Wrong guess! Try again..")

# If the player did'nt guess correctly after 3 tries
else:
    print("Sorry, you didn't guess it. The secret number was secret_number")

 # Ask if player wants to play again (while loop)
play_again = input("Do you want to play again? (yes/no): ").lower()

while play_again == "yes":
    # Restart the game
    for attempt in range(3):
        guess = int(input("Guess the secret number between one and 10: "))
        if guess == secret_number:
            print("You guessed it right! You win!")
            break
        else:
            print("Wrong guess! Try again..")

# If the player did'nt guess correctly after 3 tries
    else:
        print("Sorry, you didn't guess it. The secret number was secret_number")

 # Ask if player wants to play again (while loop)
    play_again = input("Do you want to play again? (yes/no): ").lower()

print("Thanks for playing")
```
