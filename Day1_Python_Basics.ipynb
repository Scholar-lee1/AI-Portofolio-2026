```python
def add(a, b):
  return a + b

def subtract(a, b):
  return a - b

def multiply(a, b):
  return a * b

def divide(a, b):
  return a / b

a = int(input("Value for a = "))
b = int(input("Value for b = "))

print("Basic Operators \n")

print(f"a + b =", add(a, b))
print(f"a - b =", subtract(a, b))
print(f"a - b =", multiply(a, b))
print(f"a - b =", divide(a, b))
```

```python
expenses = []
expenses.append({"item": "Transport", "amount": 1500})
expenses.append({"item": "Food", "amount": 2500})
expenses.append({"item": "Data", "amount": 1000})
expenses.append({"item": "Snacks", "amount": 500})
expenses.append({"item": "Miscellaneous", "amount": 2100})

total = sum(item["amount"] for item in expenses)
print("Total expenses today:", total)
print("All expenses:", expenses)
```

```python
#Personal expense tracker

expense = []

print("=== PERSONAL EXPENSE TRACKER ===")

# Add some expense
expense.append({"item": "Transport", "amount": 1500})
expense.append({"item": "Food", "amount": 2500})
expense.append({"item": "Data", "amount": 1000})
expense.append({"item": "Snack", "amount": 500})
expense.append({"item": "Miscellaneous", "amount": 2100})

# Show clean summary
print("Expense Summary:")
print("-" * 40)

total = 0
for exp in expense:
  print(f"{exp["item"]:<30}, #{exp["amount"]:,.0f}")
  total += exp["amount"]

print("-" * 40)
print(f"{"TOTAL SPENT":<30} #{total:,.0f}")
print("=" * 40)
```
