```python
import pandas as pd

# Creating the DataFrame
data = {
    "Name" : ["Alice", "Bob", "Charlie", "David", "Sam", "Farhan", "Mu'aadh", "Umar", "Aisha", "Maryam"],
    "Subject" : ["Math", "Science", "English", "History", "Commerce", "French", "Arabic", "Yoruba", "Physics", "Yoruba"],
    "Score" : [85, 67, 90, 72, 40, 34, 57, 76, 81, 64]
}
df = pd.DataFrame(data)

# Adding a grade column
def assign_grade(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"

df["Grade"] = df["Score"].apply(assign_grade)

# Display students who scored above 70
high_scores = df[df["Score"] > 70]

print(df)
print("\nStudents who  scored above 70")
print(high_scores)
```
