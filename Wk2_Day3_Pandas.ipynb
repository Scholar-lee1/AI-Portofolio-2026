```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/AI-2026/Small_Business_Sales_Report.xlsx'

# Read the data
df = pd.read_excel(file_path)
df.head()

# Show the first few rows
print(df.head())

# Total revenue generated
print("Total Revenue:", df["Total Revenue"].sum)

# Group by product to see the best selling product
print("\nRevenue by product:")
print(df.groupby("Product")["Total Revenue"].sum().sort_values(ascending = False))

# Filter Sales above #10,000
high_sales = df[df["Total Revenue"] > 10000]
print("\nTransactions over #10,000")
print(high_sales)

# Total quantity sold per product
print("\nQuantity sold by product")
print(df.groupby("Product")["Quantity"].sum)
```
