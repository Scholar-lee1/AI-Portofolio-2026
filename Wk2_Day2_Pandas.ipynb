```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/AI-2026/Monthly_Expenses_Tracker.xlsx'

# Read the data
df = pd.read_excel(file_path)
df.head()

# Show the first few rows
print(df.head())

# Total amount spent
print("Total Expenses:", df["Amount"].sum)

# Group by each category to see which category costs the most
print("\nExpeses by Category:")
print(df.groupby("Category")["Amount"].sum().sort_values(ascending = False))

# Count how many times each payment method was used
print("\nPayment Method Frequency")
print(df["Payment Method"].value_counts())

# Average spending per day
print("\nAverage Daily Spend")
print(df.groupby("Date")["Amount"].sum().mean())
```
