```python
# Budget plan simulation

# Problem: Help users manage their monthly expenses.

# Features:

# Create an array of expenses

# Calculate total, average, and highest expense

# Function days with expenses above average
```

```python
import numpy as np

expenses = np.array([1000, 750, 820, 950, 1200, 650, 890])
print("Total expenses:", np.sum(expenses))
print("Average expense:", np.mean(expenses))
print("Highest spending days:", np.where(expenses > expenses.mean()))
```
